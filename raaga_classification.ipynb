```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv('raga-dataset2.csv')
data.groupby("raga").size()
columns=['filename','rmse','tonal_center','chorma_cqt','chroma_stft','chroma_cens','spec_cent','spec_bw','rolloff','zcr','mfcc0','mfcc1','mfcc2','mfcc3','mfcc4','mfcc5','mfcc6','mfcc7','mfcc8','mfcc9','mfcc10','mfcc11','mfcc12','mfcc13','mfcc14','mfcc15','mfcc16','mfcc17','mfcc18','mfcc19','raga']
target = data["raga"].unique()
```

```python
data = data.sample(frac=1)
```

```python
#Encoding the Labels
raga_list = data.iloc[:, -1]
encoder = LabelEncoder()
y = encoder.fit_transform(raga_list)

#Scaling the Feature columns
scaler = StandardScaler()
X = scaler.fit_transform(np.array(data.iloc[:, :-1], dtype = float))

#Dividing data into training and Testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)
print(y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred=clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.models import Sequential
from keras.layers import LSTM, Flatten, Dense
from keras.optimizers import Adam
from scikeras.wrappers import KerasClassifier
import tensorflow as tf

# Define the function to create the LSTM model
def create_model(learning_rate=0.0001, units1=128, units2=64, dropout=0.05, recurrent_dropout=0.25):
    model = Sequential()
    model.add(LSTM(units=units1, dropout=dropout, recurrent_dropout=recurrent_dropout, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(units=units2, dropout=dropout, recurrent_dropout=recurrent_dropout, return_sequences=True))
    model.add(Flatten())
    model.add(Dense(units=96, activation="softmax"))
    adam = Adam(learning_rate=learning_rate)
    model.compile(loss="sparse_categorical_crossentropy", optimizer=adam, metrics=["accuracy"])
    return model

# Define the parameters grid for grid search
param_grid = {
    'learning_rate': [0.0001, 0.001, 0.01],
    'units1': [64, 128, 256],
    'units2': [32, 64, 128],
    'dropout': [0.2, 0.3, 0.4],
    'recurrent_dropout': [0.2, 0.3, 0.4]
}

# params={
#     'batch_size':[20,25],
#     'epochs':[50,70],
#     'model__neurons_1':[6,7],
#     'model__neurons_2':[4,3],
#     'model__activation':['relu','softmax'],
#     'model__optimizer':['adam','rmsprop'],
#     'model__dropout':[0.1,0.2]
#     }

# def create_model(neurons_1,neurons_2,activation,optimizer,dropout):
#     nn = keras.Sequential()
#     nn.add(keras.layers.Input(shape=11))       
#     nn.add(keras.layers.Dense(units=neurons_1,
#                activation=activation,kernel_initializer='glorot_uniform'))
#     nn.add(keras.layers.Dropout(rate=dropout))
#     nn.add(keras.layers.Dense(units=neurons_2,activation=activation))    
#     nn.add(keras.layers.Dropout(rate=dropout))
#     nn.add(keras.layers.Dense(units=1,activation='sigmoid'))
#     nn.compile(optimizer=optimizer,loss='binary_crossentropy',
#                    metrics=['accuracy']) 
#     return nn

model = KerasClassifier(model=create_model)

grid_search= GridSearchCV(estimator=model,param_grid=param_grid,scoring='accuracy',cv=10,
                     n_jobs=-1,return_train_score=True,verbose=0)

# Create the KerasClassifier for use in scikit-learn's GridSearchCV
model = KerasClassifier(build_fn=create_model, epochs=500, batch_size=128, verbose=0)

# Perform grid search
# grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, verbose=2)
grid_search.fit(X_train, y_train)

# Print the best parameters found by grid search
print("Best parameters found by grid search:", grid_search.best_params_)
```
